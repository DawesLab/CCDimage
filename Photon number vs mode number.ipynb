```python
# plane-wave signal
vacuum_npz = np.load("/home/photon/data/06-25-2014/11-42-08_raw.npz")
data_npz = np.load("/home/photon/data/06-25-2014/11-43-38_raw.npz")
data = data_npz['dataOut']
vacuum = vacuum_npz['dataOut']
ncount = data_npz['ncount']

# focus signal
f_data_npz = np.load("/home/photon/data/06-24-2014/14-06-21_raw.npz")
f_vacuum_npz = np.load("/home/photon/data/06-24-2014/13-56-30_raw.npz")
f_data = f_data_npz['dataOut']
f_vacuum = f_vacuum_npz['dataOut']
f_ncount = f_data_npz['ncount']
```

```python
n = np.sqrt(ncount)
print "Std dev. of LO: ", n.std()/n.mean()
print "Peak to Peak as % of LO mean: ", (n.max()-n.min())  / n.mean()
```

```python
Kp_vac = np.average(vacuum,axis=(1,2))
f_Kp_vac = np.average(f_vacuum,axis=(1,2))
```

```python
scaled_vc = np.sqrt(2.0/ncount)*(data - Kp_vac.reshape(600,1,1))
f_scaled_vc = np.sqrt(2.0/f_ncount)*(f_data - f_Kp_vac.reshape(600,1,1))
```

```python
avg_n_quick = [0.5*(np.mean(np.abs(scaled_vc[i,:,0:8].flatten('F'))))**2 for i in range(300)]
f_avg_n_quick = [0.5*(np.mean(np.abs(f_scaled_vc[i,:,15:20].flatten('F'))))**2 for i in range(300)]
```

```python
plot(avg_n_quick[150:200],".-")  # look at the modes with their amplitudes
#plot(f_avg_n_quick[150:200],".-")
xlabel("Mode number")
ylabel("Photon number")
```

```python
mode = 177
check_n = 0.5*(np.abs(scaled_vc[mode,:,:].mean(axis=0)))**2  # look at the abs of one mode averaged over all shots per round:
check_var = np.abs(scaled_vc[mode,:,:]).std(axis=0)
```

Check the average photon number for each of the rounds. Hopefully the output is stable over these longer time scales.

```python
subplot(211)
plot(check_n,".-")
ylim(ymin=0)
subplot(212)
plot(check_var,".-")
ylim(ymin=0)
```
