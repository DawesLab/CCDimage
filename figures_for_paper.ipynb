```python
vacuum_focal_raw = np.load("/home/photon/data/06-06-2014/OneKFocal/FourKShots/10-29-38_raw.npy")
```

```python
focal_raw = np.load("/home/photon/data/06-06-2014/OneKFocal/FourKShots/13-48-40_raw.npy")
```

```python
unblocked_raw = np.load("/home/photon/data/06-09-2014/Unblocked/FourKShots/14-09-15_raw.npy")
```

```python
vacuum_unblocked_raw = np.load("/home/photon/data/06-09-2014/Unblocked/FourKShots/10-47-15_raw.npy")
```

```python
unblocked_vacFFT = np.average(vacuum_unblocked_raw,axis=(1,2))
```

```python
focal_vacFFT = np.average(vacuum_focal_raw,axis=(1,2))
```

```python
VacCorrectedFocal = focal_raw - focal_vacFFT.reshape(600,1,1)
```

```python
VacCorrectedUnblocked = unblocked_raw - unblocked_vacFFT.reshape(600,1,1)
```

```python
modeA = 133
outputF = VacCorrectedFocal[modeA,:,:].flatten()  # Choose the mode to analyze
outputU = VacCorrectedUnblocked[modeA,:,:].flatten()  # Choose the mode to analyze
y_realF = np.real(outputF)
y_imagF = np.imag(outputF)
y_realU = np.real(outputU)
y_imagU = np.imag(outputU)
x_realF = []
x_realU = []
x_imagF = []
x_imagU = []
x_absF = []
x_absU = []
output_realF = []
output_realU = []
output_imagF = []
output_imagU = []
output_absF = []
output_absU = []
Pearson_realF = []
Pearson_realU = []
Pearson_imagF = []
Pearson_imagU = []
magnitudeF = []
magnitudeU = []
modes = range(0,150)
theta = []

for i in modes:
    output_realF.append(VacCorrectedFocal[i,:,:].flatten())  # Choose the mode to analyze
    output_realU.append(VacCorrectedUnblocked[i,:,:].flatten())  # Choose the mode to analyze
    x_realF.append(np.real(output_realF[i]))
    x_realU.append(np.real(output_realU[i]))
    RealPearsonF, p = sp.pearsonr(x_realF[i], y_realF)
    RealPearsonU, p = sp.pearsonr(x_realU[i], y_realU)
    Pearson_realF.append(RealPearsonF)
    Pearson_realU.append(RealPearsonU)

for i in modes:
    output_imagF.append(VacCorrectedFocal[i,:,:].flatten())  # Choose the mode to analyze
    output_imagU.append(VacCorrectedUnblocked[i,:,:].flatten())  # Choose the mode to analyze
    x_imagF.append(np.imag(output_imagF[i]))
    x_imagU.append(np.imag(output_imagU[i]))
    ImagPearsonF, p = sp.pearsonr(x_imagF[i], y_imagF)
    ImagPearsonU, p = sp.pearsonr(x_imagU[i], y_imagU)   
    Pearson_imagF.append(ImagPearsonF)
    Pearson_imagU.append(ImagPearsonU)

for i in modes:
    output_absF.append( VacCorrectedFocal[i,:,:].flatten() )  # Choose the mode to analyze
    output_absU.append( VacCorrectedUnblocked[i,:,:].flatten() )  # Choose the mode to analyze
    x_absF.append(np.abs(output_absF[i]))
    x_absU.append(np.abs(output_absU[i]))
    magnitudeF.append(np.average(x_absF[i]))
    magnitudeU.append(np.average(x_absU[i]))
    theta.append(modes[i] * (780e-9*20e6)/(600))
```

```python
plot(theta[100:150], (magnitudeF / max(magnitudeF[100:150]))[100:150],".-")
plot(theta[100:150], (magnitudeU / max(magnitudeU[100:150]))[100:150],".-.")
xlabel("$\\theta$ (mrad)")
ylabel("Scaled Amplitude")
xlim((2.5,4))
ylim((-0.1,1.1))
```

```python
plot(theta,Pearson_realF,".-")
plot(theta,Pearson_realU,".-")
xlabel("$\\theta$ (mrad)")
ylabel("Pearson Product-Moment Correlation Coefficient")
axis([-0.2,4.2,-1.1,1.1])
```
