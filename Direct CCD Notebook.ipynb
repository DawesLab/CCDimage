```python
import numpy
import sys
import matplotlib.pyplot as plt

sys.path.append("/home/photon/code/PythonForPicam")

print("Initializing PICAM library...")
from pypicam import *

ccdcam = PyPICAM()
ccdcam.configure_camera(roi=[0,210,1300,10,1,1])
ccdcam.get_temp()
```

```python
ccdcam.acquire(5)
```

```python
data = ccdcam.get_all_data()
data.shape
```

```python
for i in range(0,data.shape[2]):
    plt.plot(data[:,:,i].sum(axis=0))

plt.ylim(1.2e4,1.6e4)
```

```python
first = data[:,:,0].sum(axis=0)
second = data[:,:,1].sum(axis=0)
plt.plot(first-second)
```
